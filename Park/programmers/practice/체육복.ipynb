```python
n4=9
lost4=[1,3,5,6]
reserve4=[2,5,6]
```

```python
def solution(n, lost, reserve):
    answer = 0
    remain_lost=set(lost)-set(reserve)
    remain_reserve=set(reserve)-set(lost)
    
    condition=set([i+1 for i in remain_lost])
    remain_lost=condition-remain_reserve
    remain_reserve=set(remain_reserve)-condition
    
    condition=set([i-2 for i in remain_lost])
    remain_lost=condition-remain_reserve
    
    answer = n-len(remain_lost)
    
    return answer
```

```python
solution(n4, lost4, reserve4)
```
