```python
def decomp(data):
    num = 2
    factors = []
    while num < (data//2):
      if data%num == 0:
        factors.append(num)
        num+=1
      else:
        num+=1
    return factors

def solution(brown, yellow): 
    for y1 in decomp(yellow):
        y2=yellow/y1
        if (y1+y2)==(brown-4)/2:
          break
    answer = [int(y2+2),y1+2]
    return answer
```
